### Importación de las liberias necesarias

```python
import requests
from typing import Optional, Dict
```

### Funciones para el consumo de las apis

#### 1- Obtener una tarjeta de credito falsa de FakerAPI

```python
def obtenerTDC() -> Optional[Dict[str, str]]:
    url = "https://fakerapi.it/api/v1/credit_cards?_quantity=1"

    try:
        resp = requests.get(url)
        resp.raise_for_status() 

        datos = resp.json()
        
        tdc_datos = datos['data'][0]
        tdc = {
            "type": tdc_datos["type"], 
            "number": tdc_datos["number"], 
            "expiration": tdc_datos["expiration"], 
            "owner": tdc_datos["owner"]
        }
        return tdc
    except requests.exceptions.RequestException as e:
        print("Error en la peticion:", e)
        return None
```

### Prueba

```python
miTarjeta = obtenerTDC()
if miTarjeta is not None:
    print(f"Dueño: {miTarjeta['owner']} \n Numero: {miTarjeta['number']} \n Fecha de expiración: {miTarjeta['expiration']} \n Tipo: {miTarjeta['type']} \n")
```

### 2 - Obtener datos de un usuario de GitHub

```python
def infoUsuarioGitHub(usuario):
    url = f'https://api.github.com/users/{usuario}'
    
    try:
        resp = requests.get(url)
        
        if resp.status_code == 200:
            user_data = resp.json()
            return user_data
        else:
            print(f"Error en la peticion: {resp.status_code}")
            return None
    except Exception as e:
        print(f"Error: {e}")
        return None
```

### Prueba

```python
nombre = input("Ingresa el nombre de usuario de GitHub: ")
infoUsuario = infoUsuarioGitHub(nombre)

if infoUsuario:
    print("Información del usuario:")
    print(f"Nombre de usuario: {infoUsuario['login']}")
    print(f"ID de usuario: {infoUsuario['id']}")
    print(f"URL del perfil: {infoUsuario['html_url']}")
    print(f"Repositorios públicos: {infoUsuario['public_repos']}")
    print(f"Seguidores: {infoUsuario['followers']}")
    print(f"Siguiendo: {infoUsuario['following']}")
else:
    print("No se pudo obtener la información del usuario.")
```

### 3 - Api que obtiene publicaciones de JSON Placeholder

```python
def API_Posts():
    url = 'https://jsonplaceholder.typicode.com/posts'
    
    try:
        resp = requests.get(url)
        
        if resp.status_code == 200:
            post = resp.json()
            
            for elemento in post:
                print(f"Title: {elemento['title']}")
                print(f"Body: {elemento['body']}\n")
        else:
            print(f"Error: {resp.status_code}")
    except Exception as e:
        print(f"Error en la peticion: {e}")
```

### Prueba

```python
API_Posts()
```
